```python
from mapes_utils import EnergyMassPlot
```

```python
from siphon.simplewebservice.wyoming import WyomingUpperAir
get_upper_air_data=WyomingUpperAir.request_data 
from metpy.units import units
from datetime import datetime
import requests
import pandas as pd
import xarray as xr
```

```python

time=datetime(2018, 5, 2, 12)
station='MFL'
sounding = get_upper_air_data(time,station)
label = str(time)+' '+station

p = sounding['pressure'].values*units.hPa
T = sounding['temperature'].values*units.degC
Td = sounding['dewpoint'].values*units.degC
Z = sounding['height'].values*units.meter
u = sounding['u_wind'].values*(units.meter/units.second)
v = sounding['v_wind'].values*(units.meter/units.second)
        
```

```python
EnergyMassPlot(p,T,Td,Z,u,v,label=label)
```
